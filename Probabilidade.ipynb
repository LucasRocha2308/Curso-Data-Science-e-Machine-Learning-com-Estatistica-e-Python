**Probabilidade**

**Espaço amostral**: Todods os resultados possiveis(moeda:DUAS OPÇÕES)

**Probalidade** é a chance de um evento ocorrer

0: certo que não vai acontecer
1: certo que vai acontecer
0.5: igualmente possivel de acontecer ou de não acontecer

**Probabilidade teorica:** o que a gente espera que vai acontece ou que deveria acontecer

     P(E) = resultado/espaçoamostral

**Probabilidade Empirica:** a chance de algo acontecer baseado no que aconteceu em algum experimento.

**Lei dos grandes números (LGN):** Quanto mais experimentos eu faço, mais perto a probabilidade empirica se aproxima da probabilidade teorica.

   **Probabilidade condicional:** P(B|A)=> probabilidade de B considerando que A já ocorreu

Comprovar o teorema do limite central(TLC)

```
np.random.seed(123)
medias_amostrais = [np.mean(np.random.randint(1,7,40)) for _ in range (1000)]
plt.hist(medias_amostrais)
plt.show()
```

**Mediana**

```
np.median(medias_amostrais)
```

**Importar pandas**

```
import pandas as pd
```

**Ler DataFrame do arquivo 'weight-height.csv'**

```
from google.colab import files
uploaded = files.upload()
```

```
import io
data = pd.read_csv(io.BytesIO(uploaded['weight-height.csv']))
```

**Verificar tamanho da amostra**

```
len(data['Weight'])
```

**Gerar amostra aleatória 5 elementos**

```
np.random.seed(123)
data['Weight'].sample(n=5) #sample é amostra em inglês 
```

**Gerar 200 médias amostrais de 50 amostras cada**

```
np.random.seed(123)
medias_das_amostras = []

for _ in range(200):
  amostras = data['Weight'].sample(n=50)
  media_amostral = np.mean(amostras)
  medias_das_amostras.append(media_amostral)
```

```
len(medias_das_amostras)
```

** verificando se as medias amostrais estão em uma distribuição normal**

```
plt.hist(medias_das_amostras)
plt.show()
```

**média**

```
np.mean(medias_das_amostras)
```

**mediana**

```
np.median(medias_das_amostras)
```
